```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path= "/content/drive/MyDrive/worldbankdevelopment.csv"
df=pd.read_csv(file_path)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df.drop('date', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
print(df.columns)
```

```python
print(df.dtypes)
```

```python
environmental_columns = ['agricultural_land%', 'forest_land%', 'land_area', 'avg_precipitation', 'access_to_electricity%', 'renewvable_energy_consumption%', 'electric_power_consumption', 'CO2_emisions', 'other_greenhouse_emisions']

for col in environmental_columns:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
```

```python
economic_columns = ['trade_in_services%', 'inflation_annual%', 'real_interest_rate', 'risk_premium_on_lending', 'research_and_development_expenditure%', 'central_goverment_debt%', 'tax_revenue%', 'expense%', 'GDP_current_US']

for col in economic_columns:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
```

```python
governance_columns = ['control_of_corruption_estimate', 'goverment_effectiveness_estimate', 'political_stability_estimate', 'rule_of_law_estimate', 'regulatory_quality_estimate', 'voice_and_accountability_estimate']

for col in governance_columns:
    if col in df.columns:
        mode_val = df[col].mode()[0]  # Mode can return multiple values, take the first one
        df[col] = df[col].fillna(mode_val)
```

```python
human_development_columns = ['human_capital_index', 'doing_business', 'time_to_get_operation_license', 'statistical_performance_indicators', 'individuals_using_internet%', 'government_expenditure_on_education%', 'government_health_expenditure%', 'multidimensional_poverty_headcount_ratio%', 'gini_index']

for col in human_development_columns:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
```

```python
demographic_columns = ['birth_rate', 'death_rate', 'life_expectancy_at_birth', 'population', 'rural_population']

for col in demographic_columns:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
```

```python
df.dropna(inplace=True)
```

```python
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```
